##6. We have function $f(x)= x-1$. Give inverse function $f^{-1}$. Show both functions on the same Geogebra notebook.

to find the inverse function, Rewrite the given function using y instead of f(x).

$y=x-1$

Swap x and y:

$x=y-1$

Solve for y:

$y=x+1$

Therefore, $f^{-1}=x+1$ is the inverse function of $f(x)= x-1$
