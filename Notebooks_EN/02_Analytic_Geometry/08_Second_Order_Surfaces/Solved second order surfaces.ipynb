#$2)$

* Do the spheres with equations $x^2 + y^2 + z^2 = 1$ and $x^2 + y^2 + z^2 = 2$ have any common points?

No, the spheres do not have any common points.

The first sphere has a radius of 1, centered at the origin. The second sphere has a radius of $\sqrt2$, also centered at the origin. Since the distance between their centers is 0, which is less than the difference of their radii $(\sqrt2 - 1),$ the spheres do not intersect.

#$3)$

* What curve in space is formed by the intersection of the sphere $x^2 + y^2 + z^2 = 1$ with the sphere $(x-1)^2 + y^2 + z^2 = 1$? Find the equation of this curve.

The intersection of two spheres is typically a circle. To find the equation of this circle, we can start by subtracting the second equation from the first:

$\begin{cases} x^2 + y^2 + z^2 = 1 \\ (x-1)^2 + y^2 + z^2 = 1 \end{cases} \quad\Rightarrow\quad x^2 + y^2 + z^2-((x-1)^2 + y^2 + z^2)=1-1 \quad\Rightarrow\quad x^2 + y^2 + z^2-(x^2-1)- y^2 - z^2=0\quad\Rightarrow\quad x^2 -x^2-2x+1=0 \quad\Rightarrow\quad -2x+1=0|\times(-1) \quad\Rightarrow\quad 2x-1=0$

This equation represents a plane. The intersection of this plane with either of the two spheres will give us the desired circle.

To find the equation of the circle, we can substitute $x=\frac{1}{2}$
  into either of the sphere equations. Let's use the first one:

$(\frac{1}{2})^2 + y^2 + z^2 = 1\quad\Rightarrow\quad y^2+z^2=\frac{3}{4} $

Therefore, the intersection of the two spheres is a circle with radius $\frac{\sqrt3}{2}$ centered at the point $(\frac{1}{2},0,0)$

#$4)$

* Write the equation of the tangent plane to the paraboloid $z=(x-1)^2+y^2+1$ at point $P(1,0,1)$.

$f(x,y)=(x-1)^2+y^2+1$

$a=1,\quad b=0,\quad f(1,0)=(1-1)^2+0^2+1=1$

$\frac{\partial f}{\partial x}=2(x-1) \qquad \frac{\partial f}{\partial y}= 2y$

$\Delta f(x,y)=(2(x-1)2y) \qquad \Delta f (1,0)=(0,0)$

$a(x-x_0)+b(y-y_0)+c(z-z_0)=0 \quad\Rightarrow\quad 0(x-1)+0(y-0)+c(1-1)=0 \quad\Rightarrow\quad z=1 $

Therefore, the equation of the tangent plane to the paraboloid at point $P(1,0,1) \;is \;z = 1$
