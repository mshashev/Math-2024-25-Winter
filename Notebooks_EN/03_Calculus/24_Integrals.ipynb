###3. Calculate the area of the region bounded by the lines:
$x = 1$, $x = 2$, $y = 0$, and $y = x^2 + 1$. Show it in Geogebra.

$\int_{1}^{2}x^2+1\;dx=\int x^2\;dx+\int 1\;dx=\frac{x^3}{3}+x$
$$\left( \frac{x^3}{3}+x\right)_{1}^{2}=\left(\frac{2^3}{3}+2=\frac{8}{3}+2=\frac{8}{3}+\frac{6}{3}=\frac{14}{3}\right)-\left(\frac{1^3}{3}+1=\frac{1}{3}+1=\frac{1}{3}+\frac{3}{3}=\frac{4}{3}\right)=\frac{14}{3}-\frac{4}{3}=\frac{10}{3}$$

The area of the region bounded by the curves is $\frac{10}{3}\approx\;3.33$
