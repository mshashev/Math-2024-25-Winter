###5. Check if the function $\psi(t, x) = A \cos(\omega t + kx)$ is a solution of the second-order partial differential equation (the so-called "wave equation"), where $v = \frac{\omega}{k} = \frac{2\pi / T}{2\pi / \lambda}$:

$$
\frac{\partial^2 \psi(t, x)}{\partial t^2} - v^2 \frac{\partial^2 \psi(t, x)}{\partial x^2} = 0.
$$

1) Compute $\psi(t,x)=A\;\cos(\omega t+kx)$

Differentiate with respect to t:
$$\frac{\partial\psi(t,x)}{\partial t^2}=-A\omega\sin(\omega t+kx)$$

Again:
$$\frac{\partial^2\psi(t,x)}{\partial t^2}=-A\omega^2\cos(\omega t+kx)$$

Differentiate with respect to x:

$$\frac{\partial\psi(t,x)}{\partial x}=-Ak\sin(\omega t+kx)$$

Again:

$$\frac{\partial^2\psi(t,x)}{\partial x^2}=-Ak^2\cos(\omega t+kx)$$

2) Substitute into the wave equation:
$$-A\omega^2\cos(\omega t+kx)-v^2(-Ak^2\cos(\omega t+kx))=0 \\ -A\cos(\omega t+kx)(\omega^2-v^2k^2)=0$$

3) Simplify the condition:
Since $A\ne0$ and $\cos(\omega t+kx)\ne0$
$$\omega^2-v^2k^2=0 \;\Rightarrow\; \omega^2=v^2k^2$$

4) Substitute $v$:

$$\frac{\omega^2}{k^2}\cdot k^2=\omega^2\;\Rightarrow\; \omega^2=\omega^2$$

So the equation is true
