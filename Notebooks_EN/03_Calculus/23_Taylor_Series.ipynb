###1. Find the Taylor series and visualize obtained functions in Geogebra:

The Taylor series of a function $f(x)$ around $x=a$ is given by the formula:

$$f(x)=f(a)+f'(a)(x-a)+\frac{f''(a)}{2!}(x-a)^2+\frac{f'''ax)}{3!}(x-a)^3+...\frac{f^{n-1}(a)}{(n-1)!}(x-a)^{n-1}+...=\sum_{n=0}^{\infty}\frac{f^{(n)}(a)}{n!}(x-a)^n$$

###1) $f(x) = \cos(x)$ around $x = 0$ up to the 4th degree.

Step 1: find the derivatives:
$$f(x)=\cos(x) \\ f'(x)=-\sin(x) \\ f''(x)=-\cos(x) \\ f'''(x)=\sin(x) \\ f''''(x)=\cos(x)$$

Step 2: evaluate the function and its derivatives at x=0:

$$f(0)=\cos(0)=1 \\ f'(0)=-\sin(0)=0 \\ f''(0)=-\cos(0)=-1 \\ f'''(0)=\sin(0)=0 \\ f''''(0)=\cos(0)=1$$

Step 4: Substitute the values into the Taylor series formula:

$$f(x)=f(0)+\frac{f'(0)}{1!}x+\frac{f''(0)}{2!}x^2+\frac{f'''(0)}{3!}x^3+\frac{f''''(0)}{4!}x^4=1+\frac{0}{1!}x+\frac{-1}{2!}x^2+\frac{0}{3!}x^3+\frac{1}{4!}x^4$$

Let's simplify each component:

$ \frac{0}{1!}x=0 \\ \frac{-1}{2!}x^2=-\frac{x^2}{2} \\ \frac{0}{3!}x^3=0 \\ \frac{1}{4!}x^4=\frac{x^4}{24}$

###Finally:
###$$f(x)=1-\frac{x^2}{2}+\frac{x^4}{24}$$

###2)  $h(x) = \frac{1}{(1-x)}$ around $x = 0$ up to the 4rd degree.

Step 1: Find the derivatives:

$$h(x)=\frac{1}{(1-x)} \\ h'(x)=\frac{d}{dx}(1-x)^{-1}=-(1-x)^{-2}\cdot(-1)=(1-x)^{-2}=\frac{1}{(1-x)^2} \\ h''(x)=\frac{d}{dx}(1-x)^{-2}=-2\cdot(1-x)^{-3}\cdot(-1)=2\cdot(1-x)^{-3}=\frac{2}{(1-x)^3} \\ h'''(x)=\frac{d}{dx}(2\cdot(1-x)^{-3})=2\cdot(-3)\cdot(1-x)^{-4}\cdot(-1)=\frac{6}{(1-x)^4} \\ h''''(x)=\frac{d}{dx}(6\cdot(1-x)^{-4})=6\cdot(-4)\cdot(1-x)^{-5}\cdot(-1)=\frac{24}{(1-x)^5}$$

Step 2: Evaluate the function and its derivative at x=0:

$$h(0)=\frac{1}{1-0}=1 \\ h'(0)=\frac{1}{(1-0)^2}=1 \\ h''(0)=\frac{2}{(1-0)^3}=2 \\ h'''(0)=\frac{6}{(1-0)^4}=6 \\ h''''(x)=\frac{24}{(1-x)^5}=24$$

Step 3: Substitute the values into the Taylor series formula:

$$h(x)=h(0)+\frac{h'(0)}{1!}x+\frac{h''(0)}{2!}x^2+\frac{h'''(0)}{3!}x^3+\frac{h''''(0)}{4!}x^4= 1+\frac{1}{1!}x+\frac{2}{2!}x^2+\frac{6}{3!}x^3+\frac{24}{4!}x^4$$


Let's simplify each component:

$\frac{1}{1!}=1 \\ \frac{2}{2!}=1 \\ \frac{6}{3!}=1 \\ \frac{24}{4!}=1$

###Finally:

###$$h(x)=1+x+x^2+x^3+x^4$$

###3) $g(x) = \sin(x)$ around $x = \pi$ up to the 4rd degree.

Step 1: Find the derivatives:

$$g(x)=\sin(x) \\ g'(x)=\cos(x) \\ g''(x)=-\sin(x) \\ g'''(x)=-\cos(x) \\ g''''(x)=\sin(x)$$

Step 2: Evaluate function and its derivative at $x=\pi$:

$$g(\pi)=\sin(\pi)=0 \\ g'(\pi)=\cos(\pi)=-1 \\ g''(\pi)=-\sin(\pi)=0 \\ g'''(\pi)=-\cos(\pi)=1 \\ g''''(\pi)=\sin(\pi)=0$$

Step 3: Substitute the value into the Taylor series formula:

$$g(x)=g(\pi)+\frac{g'(\pi)}{1!}(x-\pi)+\frac{g''(\pi)}{2!}(x-\pi)^2+\frac{g'''(\pi)}{3!}(x-\pi)^3+\frac{g''''(\pi)}{4!}(x-\pi)^4=0+\frac{-1)}{1!}(x-\pi)+\frac{0}{2!}(x-\pi)^2+\frac{1}{3!}(x-\pi)^3+\frac{0}{4!}(x-\pi)^4=-(x-\pi)+\frac{1}{6}(x-\pi)^3$$

###Finally:

###$$-(x-\pi)+\frac{1}{6}(x-\pi)^3$$
